```python
df_GDP=pd.read_csv('WDIData.csv')
    # Create named reference dataframe for GDP from World Bank https://datacatalog.worldbank.org/search/dataset/0037712
```

```python
df_GDP.describe
    # Set is very large - want to reduce to GDP to compare against VG_Sales
```

```python
df_GDP=df_GDP[df_GDP['Indicator Name'].str.contains("GDP \(current US\$\)")]
    # Remove any indicator not related to GDP $ value
df_GDP
```

```python
df_GDP=pd.melt(df_GDP, id_vars=['Country Name', 'Country Code','Indicator Name','Indicator Code'],
               var_name='Year', value_name='GDP')
    # Unpivot the Range to create a column for Years
df_GDP.head()
```

```python
Area_List=pd.DataFrame({
    'Country Name':['Euro area','North America','World','Japan','Other'],
    'Area':['EU','NA','Global','JP','Other']})
    # Creating list of areas that are relevant to VG Set
```

```python
df_GDP=pd.merge(Area_List,df_GDP,on='Country Name',how='inner')
    # Merge the list of area defines with the GDP data set.
    # how=inner by default - used to truncate the table to the named areas along with the overall global figure
df_GDP.info()
    # Review the table
df_GDP = df_GDP[df_GDP.Year!='Unnamed: 66']
    # Remove redundant field 

df_GDP['Year'] = df_GDP['Year'].astype(int)
    # Convert year to int type
df_GDP = df_GDP[df_GDP.Year>=1980]
    # Remove years before 1980 - non extant in VG Set
print(df_GDP.Year.unique())
    # Print list of unique values
```

```python
df_GDP_w=df_GDP[df_GDP['Country Name']=='World']
    #Create World table in isolation for future use
df_GDP_w
```

```python
df_GDP_e=df_GDP[df_GDP['Country Name']!='World']
    #Create World table in isolation for future use
df_GDP_e
```

```python
df_GDP_e_grp=df_GDP_e.groupby('Year').sum()
    # Summarise the GDP of the areas defined in the VG_Sales section
df_GDP_e_grp
```

```python
df_GDP_we=pd.merge(df_GDP_w,df_GDP_e_grp,on='Year')
    # Merge GDP of world figures with GDP of named areas
df_GDP_we
```

```python
df_GDP_we['GDP']=df_GDP_we['GDP_x']-df_GDP_we['GDP_y']
    # Subtract GDP of named areas from global GDP to get GDP of non-named areas as a distinct field
df_GDP_we.loc[df_GDP_we['Area'] == 'Global', 'Area'] = "Other"
df_GDP_we.loc[df_GDP_we['Country Name'] == 'World', 'Country Name'] = "Other"
df_GDP_we.loc[df_GDP_we['Country Code'] == 'WLD', 'Country Code'] = "Other"
    # Rename fields per new definition
df_GDP_we=df_GDP_we[['Year','Area','GDP']]
df_GDP_we
```

```python
df_GDP=df_GDP[['Year','Area','GDP']]
df_GDP
```

```python
df_GDP = pd.concat([df_GDP_we, df_GDP],  ignore_index = True)
df_GDP
```

```python
df_GDP=df_GDP[['Area', 'Year', 'GDP']]
df_GDP
```

```python
df_GDP_pivot = df_GDP.pivot(index="Year",columns="Area", values="GDP")
df_GDP_pivot = df_GDP_pivot.rename(columns = {'EU':'EU GDP','Global':'Global GDP','JP':'JP GDP','NA':'NA GDP','Other':'Other GDP'})
df_GDP_pivot
```

```python
df_VG_Sales_grp=df_VG_Sales.groupby('Year_of_Release').sum()
df_VG_Sales_grp=df_VG_Sales_grp.drop(['Critic_Score','Critic_Count','User_Score','User_Count'], axis = 1)
df_VG_Sales_grp['User_Weight']=10*df_VG_Sales_grp['User_Weight']/df_VG_Sales_grp['Global_Sales']
df_VG_Sales_grp['Critic_Weight']=df_VG_Sales_grp['Critic_Weight']/df_VG_Sales_grp['Global_Sales']
df_VG_Sales_grp
```

```python
df_VG_Sales_grp = df_VG_Sales_grp.rename(index = {'Year_of_Release':'Year'})
    # Changing label on Year_of_Release to Year to simplify later join
df_VG_Sales_grp.index.name ='Year'
df_VG_Sales_grp.iloc[:,0:5] = df_VG_Sales_grp.iloc[:,0:5].mul(1000000)
    # Returning units sold count to millions of untis - not editing the aggregate on ratings weighted
df_VG_Sales_grp_S = df_VG_Sales_grp.rename_axis('Year').reset_index()
df_VG_Sales_grp_S
```

```python
df_VG_Sales_grp_S=pd.merge(df_GDP_pivot,df_VG_Sales_grp_S,on='Year',how='inner')
    # Joining data on GDP to date on unit Sales
df_VG_Sales_grp_S
    # Reviewing sales
```

```python
df_VG_Sales_grp_S['Global_Ratio'] = df_VG_Sales_grp_S['Global_Sales'].div(df_VG_Sales_grp_S['Global GDP'])
df_VG_Sales_grp_S['Other_Ratio'] = df_VG_Sales_grp_S['Other_Sales'].div(df_VG_Sales_grp_S['Other GDP'])
df_VG_Sales_grp_S['EU_Ratio'] = df_VG_Sales_grp_S['EU_Sales'].div(df_VG_Sales_grp_S['EU GDP'])
df_VG_Sales_grp_S['NA_Ratio'] = df_VG_Sales_grp_S['NA_Sales'].div(df_VG_Sales_grp_S['NA GDP'])
df_VG_Sales_grp_S['JP_Ratio'] = df_VG_Sales_grp_S['JP_Sales'].div(df_VG_Sales_grp_S['JP GDP'])
df_VG_Sales_grp_S
```

```python
    # Creating plots of units shipped/sold over time by region

sb.lmplot(data=df_VG_Sales_grp_S, x='Year', y='JP_Ratio',order=4, height=4, aspect=4)
plt.title('Japan - Ratio of Video Game Units Sold to GDP')
plt.ylim(0, None)
plt.xlim(1980, 2016)
plt.show()

sb.lmplot(data=df_VG_Sales_grp_S, x='Year', y='NA_Ratio',order=4, height=4, aspect=4)
plt.title('North America - Ratio of Video Game Units Sold to GDP')
plt.ylim(0, None)
plt.xlim(1980, 2016)
plt.show()

sb.lmplot(data=df_VG_Sales_grp_S, x='Year', y='EU_Ratio',order=3, height=4, aspect=4)
plt.title('Europe - Ratio of Video Game Units Sold to GDP')
plt.ylim(0, None)
plt.xlim(1980, 2016)
plt.show()

sb.lmplot(data=df_VG_Sales_grp_S, x='Year', y='Other_Ratio',order=3, height=4, aspect=4)
plt.title('Other World - Ratio of Video Game Units Sold to GDP')
plt.ylim(0, None)
plt.xlim(1980, 2016)
plt.show()

sb.lmplot(data=df_VG_Sales_grp_S, x='Year', y='Global_Ratio',order=3, height=4, aspect=4)
plt.title('Global - Ratio of Video Game Units Sold to GDP')
plt.ylim(0, None)
plt.xlim(1980, 2016)
plt.show()
```
